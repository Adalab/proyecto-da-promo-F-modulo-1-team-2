```python
import random

print("""
    - BIENVENIDO AL JUEGO DE PIEDRA, PAPEL O TIJERA. 
    - Vas a jugar contra mí, ¡mua ha ha haaaa!
    - Solo tienes que elegir una de estas tres opciones: piedra, papel o tijera.
      Recuerda: piedra aplasta a tijera, papel envuelve a piedra y tijera corta a papel.
    - Quien gane tres veces gana la partida. ¡Suerte! Las vas a necesitar ¡mua ha ha haaaa!
""")

puntos_jugador = 0
puntos_pc = 0
# Añadir reglas del juego y añadir contadores visibles (yo llevo tanto y tú tanto)

while puntos_jugador < 3 and puntos_pc < 3:
    opciones = ['piedra','papel','tijera']

    tu_eleccion = input("Escribe que quieres: piedra, papel o tijera.").lower()
    while tu_eleccion not in opciones:
        print("No has introducido ninguna opción de las dadas.")
        tu_eleccion = input("¿Qué quieres: piedra, papel o tijera.?").lower()

    pc_eleccion = opciones[random.randint(0,2)] 
    # Elección al azar del ordenador usando los índices para acceder a la lista de cartas.
    
    print(f"\t\tMi elección es:{pc_eleccion}")
    print(f"\t\tTu elección es:{tu_eleccion}")

    # Sacamos los índices de la elección del usuario y la del ordenador.
    indice_tu_palabra = opciones.index(tu_eleccion)
    indice_pc_palabra = opciones.index(pc_eleccion)

    # Comenzamos las comparaciones
    if indice_tu_palabra == 0 and indice_pc_palabra == 2: # Piedra gana a papel aunque es 0 y 2
        puntos_jugador += 1
        print(f"""\nA ver... {tu_eleccion} gana a {pc_eleccion}\n
            ¡ME HAS GANADO!""", "\N{confounded face}\n")    
    else:
        if indice_tu_palabra < indice_pc_palabra:
            puntos_pc += 1
            print(f"""\nA ver... {pc_eleccion} gana a {tu_eleccion}\n
                ¡TE HE GANADO!""","\N{relieved face}\n")
        elif indice_tu_palabra > indice_pc_palabra:
            puntos_jugador += 1
            print(f"""\nA ver... {tu_eleccion} gana a {pc_eleccion}\n
                ¡ME HAS GANADO!""","\N{confounded face}\n")
        else:
            print("\n\t\tIguales, empatamos \N{neutral face} \n")
    if puntos_pc == puntos_jugador == 2:
        print("\t\tEstamos empatados. ¡¡¡Quien acierte gana el juego!!!\N{melting face}\n")
if puntos_pc == 3:
    print("\n¡HE GANADO EL JUEGO!")
else:
    print("\n¡ENHORABUENA! ¡HAS GANADO EL JUEGO!\n")
    print("""
                                                      
{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}
{} ___ ___  _______  _______      _______  ______  _______      _______ {}
{}|   |   ||       ||   |   |    |   _   ||   __ \|    ___|    |   _   |{}
{} \     / |   -   ||   |   |    |       ||      <|    ___|    |       |{}
{}  |___|  |_______||_______|    |___|___||___|__||_______|    |___|___|{}
{}                                                                      {}
{} ________  _______  _______  _______  _______  ______                 {}
{}|  |  |  ||_     _||    |  ||    |  ||    ___||   __ \                {}
{}|  |  |  | _|   |_ |       ||       ||    ___||      <                {}
{}|________||_______||__|____||__|____||_______||___|__|                {}
{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}""")

print("Bye, bye, my friend! \N{crying face}")
```
