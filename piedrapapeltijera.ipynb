```python
import random

points_player = 0
points_computer = 0
# Añadir reglas del juego y añadir contadores visibles (yo llevo tanto y tú tanto)

while points_player < 3 and points_computer < 3:
    cards = ['piedra','papel','tijera']

    your_election = input("Escribe que quieres: piedra, papel o tijera.").lower()
    while your_election not in cards:
        print("No has introducido ninguna opción de las dadas.")
        your_election = input("¿Qué quieres: piedra, papel o tijera.?").lower()

    computer_election = cards[random.randint(0,2)] 
    # Elección al azar del ordenador usando los índices para acceder a la lista de cartas.
    
    print(f"\t\tMi elección es:{computer_election}")
    print(f"\t\tTu elección es:{your_election}")

    # Sacamos los índices de la elección del usuario y la del ordenador.
    your_word_indice = cards.index(your_election)
    computer_word_indice = cards.index(computer_election)

    # Comenzamos las comparaciones
    if your_word_indice == 0 and computer_word_indice == 2: # Piedra gana a papel aunque es 0 y 2
        points_player += 1
        print(f"""A ver... {your_election} gana a {computer_election}
            ¡ME HAS GANADO!""")      
    else:
        if your_word_indice < computer_word_indice:
            points_computer += 1
            print(f"""A ver... {computer_election} gana a {your_election}
                ¡TE HE GANADO!""")
        elif your_word_indice > computer_word_indice:
            points_player += 1
            print(f"""A ver... {your_election} gana a {computer_election}
                ¡ME HAS GANADO!""")
        else:
            print("\t\tIguales, empatamos")
    if points_computer == points_player == 2:
        print("\t\tEstamos empatados. ¡¡¡Quien acierte gana el juego!!!")
if points_computer == 3:
    print("¡HE GANADO EL JUEGO!")
else:
    print("¡ENHORABUENA! ¡HAS GANADO EL JUEGO!")

print("Bye, bye, my friend!")
```
